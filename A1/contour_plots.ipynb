```python
import numpy as np
import matplotlib.pyplot as plt
```

Consider the function $f: \R^2 \to \R$: 

$$ f(x_1,x_2) = -2x_1^2 + x_1 x_2^2 + 4x_1^4 $$

```python
def func(x1,x2):
    return np.exp(-2*x1**2 + x1*x2**2 + 4*x1**4)
```

```python
x1, x2 = np.meshgrid(np.linspace(-0.7,0.7,100), np.linspace(-0.6,0.6,100))

print(x1)
```

```python
# surface plot
fig, ax = plt.subplots(1,1,figsize=(10,5))
ax = plt.axes(projection='3d')
ax.plot_surface(x1, x2, func(x1,x2), cmap='viridis')
plt.show()
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,5))
cs = ax.contour(x1,x2,func(x1,x2),levels=50)
ax.clabel(cs, inline=1, fontsize=10)
plt.show()
```

```python
# reduce range just for demo of stationary point
x1, x2 = np.meshgrid(np.linspace(-0.1,0.1,100), np.linspace(-0.6,0.6,100))

fig, ax = plt.subplots(1,1,figsize=(10,5))
cs = ax.contour(x1,x2,func(x1,x2),levels=50)
ax.clabel(cs, inline=1, fontsize=10)
plt.show()
```

```python
# add markers to stationary points
x1, x2 = np.meshgrid(np.linspace(-0.7,0.7,100), np.linspace(-0.6,0.6,100))

fig, ax = plt.subplots(1,1,figsize=(10,5))
cs = ax.contour(x1,x2,func(x1,x2),levels=40)
ax.clabel(cs, inline=1, fontsize=10)
ax.plot(0,0,'*',label='saddle point')
ax.plot(-0.5,0,'*',label='saddle point')
ax.plot(0.5,0,'*',label='local minimum')
ax.legend()
plt.show()
```
